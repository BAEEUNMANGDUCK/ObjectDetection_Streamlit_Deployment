```python
import plotly.graph_objects as go
import plotly.io as pio

# Google Colab에서는 반드시 렌더러를 "colab"으로 설정
pio.renderers.default = "colab"
```

```python
# plotly.graph_objects 에서는 x와 값을 지정해줘야 함
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x= [1, 2, 3], y=[1, 2, 3]
    )
)
fig.show()
```

```python
# 라인 그래프 테스트
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 2), columns = ['A', 'B'])
df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=df.index, y=df['A']
    )
)



# fig.update_layout() -> layout 업데이트

fig.update_layout(
    {
        "title": {
            "text": "Graph with go.Scatter",
            "font": {
                "size": 15
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "random number"
        },
        "yaxis": {
            "title": "A"
        }
    }
)

fig.show()
```

```python
# 같은 창 안에서 그래프 두개 그리고 싶을 때
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=df.index, y=df['A'], mode= 'lines', name= 'A'
    )
)

fig.add_trace(
    go.Scatter(
        x=df.index, y=df['B'], mode="lines+markers+text", name='B', text=df.index, textposition='top center'
    )
)
fig.show()
```
