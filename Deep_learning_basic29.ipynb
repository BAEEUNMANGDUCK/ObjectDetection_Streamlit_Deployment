#### 선형 회귀
  - ex: 시험 공부를 하는 시간이 길면, 성적이 잘 나온다
  - 특정 변수(정보)에 따라, 또다른 특정 변수 값이 영향을 받음
  - 성적을 변하게 하는 정보 요소를 독립변수(x)라 하고, 독립변수 값에 따라, 종속적으로 변하는 변수를 종속변수(y)라고 함.
  - 선형 회귀는 한 개 이상의 독립변수와 종속 변수간의 관계를 선형 모델로 구현

#### 단순 선형 회귀
 - 독립변수가 하나인 경우
 - 단순 선형회귀식: y = wx + b
  - 독립변수: x
  - 가중치 w, 편향 b

#### 다중 선형 회귀
 - 독립변수가 여러개인 경우
 - 다중 선형 회귀식: y = w1x1 + w2x2 ... + wnxn + b
 - 독립변수: x1, x2, x3
 - 가중치: w1, w2, w3

#### 선형회귀가 예측을 하는 방법(예측 문제 분석)
- 독립변수: 수학 공부 시간
- 종속변수: 수학 성적

####🔧 최소제곱법은 "해를 직접 계산하는 공식"
- 선형 회귀 같은 간단한 모델에서는 해를 직접 수식으로 구할 수 있습니다.

- 빠르고 정확하지만, 다음과 같은 경우에 문제가 생깁니다:

- 데이터가 매우 많거나 차원이 높을 때: 행렬 계산이 너무 느리고 메모리를 많이 사용함

- 비선형 모델이나 딥러닝에는 적용 불가 (복잡한 구조에 닫힌 해가 없음)

- 정규화, 제약조건, 손실 함수가 제곱이 아닐 때는 적용할 수 없음

####🔁 경사하강법은 "해를 점점 찾아가는 반복 알고리즘"
- 초기 값에서 시작해서, 손실 함수를 줄이는 방향으로 조금씩 파라미터를 업데이트함

- 대부분 비선형, 고차원, 비선형 활성화 함수, 복잡한 손실을 쓰기 때문에 수식으로 해를 구할 수 없기에 경사하강법을 사용

- 미분을 통해 순간 기울기를 계산하고, learning rate 만큼 반영하여, 조금씩 최소값을 찾아가는 형태

- 순간 기울기가 음수이면 가중치에 양수가 더해지고, 양수면 음수가 더해지며, 그 비율이 학습률에 의해 조정됨

- learning rate를 너무 높게 잡으면 최소값에 근접하지 못하고 발산하며, 너무 작으면 학습에 오랜 시간이 걸림

#### 텐서를 정의하는 용어
- 축(axis)의 개수(rank): 1D 텐서(Vector)는 1개의 축이 있고, 2D 텐서(Matrix)는 2개의 축이 있음
  - pytorch에서는 텐서객체.dim()으로 축의 개수를 확인할 수 있음
  - 축의 개수를 다른 말로 rank라고 함

- 크기(shape 또는 size): 텐서의 각 축(axis)에 따라 얼마나 많은 차원이 있는지를 나타내는 값으로, 파이썬의 튜플 구조로 표현
- 용어가 헷갈리므로, shape,size를 기반으로 실제 구조를 확인하는 편이 좋음
  - ex: 2D 텐서는: 축이 두개 있다는 의미 (matrix)
  - ex: 3D 벡터는 벡터이므로 축이 1개이지만 3개의 원소를 가지고 있다는 의미임

#### Tensor 연산 (브로드캐스팅)
- Tensor 간 연산이 broadcasting을 지원하면, Tensor는 동일한 차원으로 자동 확장됨

#### broadcasting 이해
- 두 tensor 중 하나의 차원의 원소 수가 1이거나, 존재하지 않는 경우, broadcasting이 일어나면, 해당 차원이 확대됨
image.png


image.png

```python
import torch
data1 = torch.FloatTensor([
    [1],
    [2],
    [3]
])

data2 = torch.FloatTensor([1, 2, 3])

data3 = data1 + data2
print(data1.shape, data2.shape, data3.shape)
print(data3)
```

#### broadcasting 조건
- 두 tensor 간의 브로드캐스팅이 가능하기 위한 조건이 있음
  - 두 tensor에 있는 각 차원에 대해, 끝쪽 차원부터 다음조건을 비교하며, 앞쪽방향으로 진행
    - 각 차원의 원소 수가 똑같거나'
    - 둘 중 하나의 차원의 원소 수는 1이거나, 존재하지 않는 경우
- 주요 브로드캐스팅 가능 케이스를 broadcastable 이라고 함

#### boolean indexing
- 조건 필터링과 검색을 동시에 할 수 있음

```python
data1 = torch.FloatTensor([
    [1, 2, 3],
    [5, 5, 5]
])

data2 = data1 > 3
print(data2)
print(data1[data2])
print(data1[(data1 > 3) & (data1 < 5)])
```

#### 텐서 복사: 텐서객체.clone().detach()
- clone(): 기존 텐서객체의 내용을 복사한 텐서 생성 (깊은 복사)
- detach(): 기존 텐서객체 그래프에서 분리된 텐서로 구성

```python
data1 = torch.FloatTensor([
    [1, 2, 3],
    [4, 5, 6]
])
print(data1)
data2 = data1[:2, :2]
print(data2)
```

```python
data2[1, 1] = 4
print(data2)
print(data1)
```

```python
data1 = torch.FloatTensor([
    [1, 2, 3],
    [4, 5, 6]
])
print(data1)
data2 = data1[:2, :2].clone().detach()
print(data2)
data2[1, 1] = 4
print(data2)
print(data1)
```

#### 조건에 맞는 값 인덱싱

#### 퍼셉트론(perceptron)

뉴런의 이해
- 신경세포 1개를 의미하며, 인간의 뇌는 약 1000억개의 뉴런으로 구성됨
- 뉴런은 크게 세가지 부분으로 나뉨
  - 신경세포체: 핵이 있는 세포 부분
  - 가지돌기: 다른 세포에서 신호를 받는 부분
  - 축삭돌기: 다른 세포에 신호를 주는 부분
  - 시냅스: 돌기 사이에서 다른 뉴런에 신호를 전달하는 부분
-가지돌기에 자극을 받게 되면, 신경세포체를 통해 축삭돌기에 화확물질이 전달되어 전이 발생
- 전이의 정도가 특정 임계값을 넘으면, 시냅스를 통해 다른 뉴런으로 전달


#### 인공신경망(ANN)
- 워렌 맥컬록(Warren McCulloch)과 월터 피츠(Walter Pitts) 이 1943년 논문 발표
  - 뉴런의 작용을 0과 1로 이루어진 이진법 논리 회로 모델로 구성
  - 이런 모델을 그물망처럼 연결하면, 인간의 뇌처럼 기계를 작동시킬 수 있을 것이라 주장
- 프랑크 로젠블라트가 이 이론을 실제 장치로 구현
- 이를 **퍼셉트론**이라 함.
  - x1, x2, x3...xn을 각각의 입력값으로 받음
  -  x1 * w1 + .... b 각각을 가중치와 곱함
  - 특정 임계값을 넘으면 1 아니면 0을 출력
  - 실제 결과와 오차가 생기면, 가중치를 업데이트
    - 오차 기반 가중치 업데이트를 할 수 있는 구조로 인해, 최초로 학습이 가능해짐

#### nn.Module 기반, nn.Parameter 등록하기
- 학습 대상이 되는 텐서는 해당 모듈에 연결된 Parameter로 등록해야 함
  - 이를 통해 특정 모듈에서 학습 처리시 필요한 작업을 알아서 해주도록 구성되어 있음
  - 모듈 내에서 학습 대상이 되는 텐서들은 __init__에서 nn.Parameter()으로 등록해줘야 함
    - nn.Parameter(텐서, requires_grad=True)
      - requires_grad 는 default True이며, 파라미터가 gradient 방식으로 계산되는 케이스 의미
      - 내부적으로 gradient 방식에서 필요한 미분 연산을 위해, 파이토치는 동적으로 그래프를 구성하므로, 이런 구성이 필요함

```python
import torch
import torch.nn as nn

class NeuralNetwork(nn.Module):

  def __init__(self, input_dim, output_dim):
    super().__init__()
    self.W = nn.Parameter(torch.FloatTensor(input_dim, output_dim))
    self.b = nn.Parameter(torch.FloatTensor(output_dim))

  def forward(self, x):

    y = torch.matmul(x, self.W)  + self.b
    return y
```

```python
x = torch.FloatTensor(4)
mylinear = NeuralNetwork(4, 3)
y = mylinear(x)
print(y, y.shape)
for param in mylinear.parameters():
  print(param)
```

#### 기본 Optimizer
- 딥러닝의 학습은 손실값이 가장 작은 모델을 만드는 것
- 즉, 학습은 손실함수의 최소값을 찾아가는 과정이며, 이를 Optimization 이라고 함
  - 최적화를 수행하는 알고리즘을 Optimizer 라고 함
- 대표적인 Optimizer 는 Gradient Descent 이며 이외에 다양하게 있음

#### BGD 와 SGD
- BGD: 배치 경사 하강법
  - 전체 데이터셋을 한번에 학습
  - 메모리 제약 사항, 속도 성 문제 발생

- SGD: 확률 경사 하강법
  - 전체 학습 데이터 중 특정 크기만큼을 임의로 선택하여 학습

- 미니배치
  - epoch: 전체 데이터셋이 한번 최적화를 수행한 것
  - iteration: 하나의 배치(일정 크기의 데이터 사이즈)가 한번 최적화를 수행한 것
  - 미니배치
    - 전체 데이터셋이 1000개 있을 때
    - 이를 200개의 배치 사이즈로 분리하면, 200개의 서브 데이터셋을 미니배치라고 함
    - 1 epoch를 위해서는 5 iteration 이 필요함

#### BGD
  - 이때의 배치는 **전체 데이터셋**을 의미함
  - 전체 데이터셋을 기반으로 학습하므로, 모델 파라미터 업데이트에 상대적으로 오래 걸림
  - 전체 데이터셋으로 학습시키기 때문에, 수렴되는 정도가 일정
  - 수렴되는 정도가 안정적이기 때문에, local minima 상태가 되면 빠져나오기 어려움



#### local minima 와 global minima
- local minima 는 지역 최소점으로, 주변 cost 함수값 중 최소값을 가지는 지점 의미
- global minima 는 전역 최소점으로, 전체 cost 함수값이 최소값을 가지는 지점 의미

#### SGD (Stochastic Gradient Descent)
- BGD의 단점을 보완하기 위해 고안
  - 전체 데이터 중 **랜덤하게** 선택된 **k개의 샘플 데이터**를 이용하여 학습시키는 경사 하강법
  - 본래는 전체 데이터 중 1개의 샘플 데이터만 사용해서, 학습시키는 경사하강법
    - 이 때, epoch를 1000회 수애하면, 1 epoch 당 1개의 샘플 데이터를 가지고 한번 학습 및 가중치 업데이트
    - 따라서, 총 1000회의 학습 및 가중치 업데이트
  - 1개의 샘플 데이터가 아닌, 랜덤하게 k개의 샘플 데이터를 사용하여 학습시키는 경사 하강법은 Mini-Batch Gradient Descent라고 함

#### 자동이분(torch.autograd)
- Pytorch의 autograd는 신경망 훈련을 지원하는 자동 미분 기능
- torch.autograd 동작 방법
  - 텐서에 .requires_grad 속성을 True로 설정하면, 이후의 텐서 모든 연산들을 추적함
  - 텐서.backward()를 호출하면, 연산에 연결된 각 텐서들의 미분 값을 계산하여, 각 텐서객체.grad에 저장
    - requires_grad()는 연결된 텐서로부터의 계산된 자동 미분 값을 다시 현 텐서부터 시작하도록 만듬

#### 신경망 동작 이해
- 모델 및 데이터 생성
- forward pass로 입력 데이터를 모델에 넣어서 예측값 계산
- 예측값과 실제값의 차이를 loss function으로 계산
- backward pass로 각 모델 파라미터를 loss 값 기반하여 미분하여 저장
- optimizer로 모델 파라미터의 최적값을 찾기 위해 파라미터 값 업데이트

#### 파라미터 업데이트

```python
import torch
import torch.nn.functional as F

# 예시용 초기값
x = torch.ones(4)
y = torch.zeros(3)
W = torch.rand(4, 3, requires_grad=True)
b = torch.rand(3, requires_grad=True)

threshold = 0.1
learning_rate = 0.1
iteration_num = 0

z = torch.matmul(x, W) + b
loss = F.mse_loss(z, y)
loss.backward()

while loss > threshold:
    iteration_num += 1

    with torch.no_grad():
        W -= learning_rate * W.grad
        b -= learning_rate * b.grad

    W = W.detach().clone().requires_grad_(True)
    b = b.detach().clone().requires_grad_(True)

    z = torch.matmul(x, W) + b
    loss = F.mse_loss(z, y)
    loss.backward()

    print(iteration_num, loss.item(), z.mean().item(), y.mean().item())

print(iteration_num + 1, loss.item(), z.mean().item(), y.mean().item())
```

#### Optimizer와 경사하강법
- 위의 과정은 수동으로 가중치 업데이트
- 자동으로 최적화하는 방법
  - 초기화(모델 파라미터 리스트와 learning rate를 넣어줌)
    - optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
  - 학습단계(반복 수행하며 최적화하는 단계)
    - optimizer.zero_grad()를 호출하여 모델 파라미터 미분값을 0으로 초기화해줌
    - loss.backward()를 호출하여 backward pass()를 계산
    - optimzer.step()을 호출하여 역전파 단게에서 계산된 미분값 기반으로 모델 파라미터 업데이트


#### optimizer.zero_grad()
- 모델 파라미터의 미분값을 초기화해주지 않으면, 이어지는 모델 파라미터 연산 시, 해당 연산에 대한 미분값에 기존 미분값을 더해줌
- 따라서 해당 연산에 대한 미분값이 필요한 경우, optimizer.zero_grad()를 통해 기존 미분값을 0으로 초기화해줘야 함.

#### 자연상수(e) 이해를 위한 로그와 로그 함수

  - 일정기간 동안 100% 복리이자율을 주되, 횟수를 연속적으로 무한대로 늘리면?
