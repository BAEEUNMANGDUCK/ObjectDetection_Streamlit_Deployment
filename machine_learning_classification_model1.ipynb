#### 1. K-Nearest Neighbor(KNN)

-k값(거리)을 설정한 후, k 거리 안에 들어온 기존 데이터(훈련 데이터)의 타입(분류)이 다수인쪽으로 예측하는 기법
- 장점: 단순함
- 단점: 적절한 K 값 설정에 어려움이 있고, 데이터와 feature수가 많아지면 예측 시간이 오래 걸림
- 분류에도 쓰이지만, 군집화에서 더 많이 쓰이는 머신러닝 기법

#### 로지스틱 회귀
- Logistic Regression은 회귀모델이 아니라, 분류 모델임
- Linear Classification 과 달리 'S'자 모양의 곡선으로 계산하는 시그모이드 함수를 사용해서, 분류하는 모델
  - 현실세계에서 특정현상은 특정변수에 대한 확률값이 선형이 아닌 .S-커브 형태를 따르는 경우가 많음
- 로지스틱 회귀는 정확히 0 또는 1을 예측하는 대신에 확률(0~1 사이의 값)을 생성해서 0과 1을 분류하는 예측 모델

#### 서포트 벡터 머신 (SVM)
  - 분류를 위해, 분류의 경계가 되는 경계선을 작성하여, 분류 실행
    - margin: 결정선과 가장 가까운 양 옆의 데이터 간의 거리
    - 결정선과 가장 가까운 양 옆의 데이터를 Support Vector라고 함
    - SVM은 margin을 최대화하는 결정선을 구하는 알고리즘
  - sklearn 라이브러리에서는 svc로 사용됨
